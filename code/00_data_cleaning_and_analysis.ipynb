Data Cleaning/Preprocessing

```python
##Drop irrelevant rows
pre_clean= pre.drop(0)
pre_clean=pre_clean.drop(1)

pre_clean

pre_clean.to_csv("precleaned_df")
```

```python
#Q6_correct answers column
pre_clean['Q6_correct']= np.where(pre_clean['Q6']== 'Anxiety, Depression, & Trauma-Related Disorders', 1, 0)
pre_clean.loc[(pre_clean['Q6'].isna()) & (pre_clean['Q6_correct'] == 0), 'Q6_correct'] = np.nan


#Q37_correct answers column
pre_clean['Q37_correct']= np.where(pre_clean['Q37']== 'Talking to inanimate objects', 1, 0)
pre_clean.loc[(pre_clean['Q37'].isna()) & (pre_clean['Q37_correct'] == 0), 'Q37_correct'] = np.nan


#Q13_correct answers column

#This one was multiple choice but only one answer is correct so basically everone got it wrong

pre_clean['Q13_correct']= np.where(pre_clean['Q13']== 'Defer to the person’s caregiver', 1, 0)
pre_clean.loc[(pre_clean['Q13'].isna()) & (pre_clean['Q13_correct'] == 0), 'Q13_correct'] = np.nan

#Q22_correct answers column

pre_clean['Q22_correct']= np.where(pre_clean['Q22']== 'Combination treatment, including pharmacotherapy/manualized psychotherapies', 1, 0)
pre_clean.loc[(pre_clean['Q22'].isna()) & (pre_clean['Q22_correct'] == 0), 'Q22_correct'] = np.nan


pre_clean
```

Preliminary Data Analysis


```python
#Importing pandas and NumPy
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import random
import re
```

```python
#Means of Correct Scores for Multiple Choice Answers


correct_columns= ['Q6_correct','Q37_correct', 'Q13_correct', 'Q22_correct']  


correct_percentages={}
for col in correct_columns:
    percentage_correct = pre_clean[col].mean() *100
    correct_percentages[col] = percentage_correct

summary_df = pd.DataFrame.from_dict(correct_percentages, orient='index', columns=['Percentage correct'])
summary_df
```
